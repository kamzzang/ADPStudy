```python
import numpy as np

x = np.linspace(-5, 5, 101)
x
```

```python
y = (1 / np.sqrt(2*np.pi) * np.exp(-x**2/2))
y
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Normal distribution without scipy')
plt.legend(['N(0,1)'])
plt.show()
```

## scipy.stats 이용

```python
import scipy.stats as stats

y1 = stats.norm(0, 1).pdf(x)

plt.figure(figsize=(6,4))
plt.plot(x, y1, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Normal distribution with scipy.stats')
plt.legend(['N(0,1)'])
plt.show()
```

```python
legend = []

def norm(mu, var):
    legend.append("N(" + str(mu) + ", " + str(var) + ")")
    return stats.norm(mu, var).pdf(x)

plt.figure(figsize=(6,4))

for i in np.arange(4) + 1:
    plt.plot(x, norm(0, i))
    
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Normal distribution with scipy.stats')
plt.legend(legend)
plt.show()
```
